```python
from datasets import load_dataset
import os
from PIL import Image
import numpy as np

# 데이터셋 로드
dataset = load_dataset("naver-clova-ix/cord-v2")

# 이미지를 저장할 디렉토리 생성
save_dir = "eng_receipt"
os.makedirs(save_dir, exist_ok=True)

# 이미지 저장 함수
def save_images(split_name):
    for i, item in enumerate(dataset[split_name]):
        # 이미지 데이터를 array로 로드
        image_array = np.array(item["image"])
        image = Image.fromarray(image_array)  # array를 PIL Image로 변환

        image_path = os.path.join(save_dir, f"{split_name}_{i}.jpg")
        image.save(image_path)
        print(f"Saved {image_path}")

# train, validation, test split 이미지 저장
for split in ["train", "validation", "test"]:
    save_images(split)
```
